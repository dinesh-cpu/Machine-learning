## Approach 4: Using Imputer

##### strategies to fill the missing values: 
- mean
- median
- most_frequent

```python
from sklearn.impute import SimpleImputer

imputer_mean = SimpleImputer(missing_values = np.nan, strategy = "mean")
imputer_median = SimpleImputer(missing_values = np.nan, strategy = "median")
```

```python
imputed_data_mean = imputer_mean.fit(data[cols])
imputed_data_mean = imputer_mean.transform(data[cols])
```

```python
imputed_data_median = imputer_median.fit_transform(data[cols])
```

```python
imputed_data_mean
```

```python
imputed_data_median
```

```python
pd.DataFrame(data = imputed_data_median, columns = cols)
```
