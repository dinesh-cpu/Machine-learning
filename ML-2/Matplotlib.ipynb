### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Line Plot

```python
x = [2, 5, 5, 6, 7, 8]
y = [3, 7, 7.2, 4, 12, 5]

plt.plot(x, y)
plt.plot(x, y, '^')
plt.title('Line Graph')
plt.xlabel("x-axis values")
plt.ylabel("y-axis values")
plt.show()
```

```python
x = np.arange(0, 3*np.pi, 0.1)
```

```python
y = np.cos(np.arange(0, 3*np.pi, 0.1))
```

```python
plt.plot(x, y)
plt.title('Line Graph showing sine wave')
plt.xlabel("x-axis values")
plt.ylabel("y-axis values")
plt.show()
```

```python
plt.plot(x, y)
plt.title('Line Graph showing cosine wave')
plt.xlabel("x-axis values")
plt.ylabel("y-axis values")
plt.show()
```

### Scatter Plot

```python
x = [2, 5, 5, 6, 7, 8]
y = [3, 7, 7.2, 4, 12, 5]

plt.scatter(x, y)
plt.title('Line Graph')
plt.xlabel("x-axis values")
plt.ylabel("y-axis values")
plt.show()
```

## Comparison

```python
# team a
x1 = [2 ,5, 6, 7, 8]
y1 = [3 ,7, 4, 12 ,5]

# team b
x2 = [4 ,5, 8, 9, 8]
y2 = [3 ,5, 7, 12 ,6]

plt.plot(x1, y1, color = 'g', linewidth = 4)
plt.plot(x2, y2, color = 'r', linewidth = 2)

plt.annotate(xy = [2, 3], s = "Team_A")
plt.annotate(xy = [4, 4], s = "Team_B")

plt.legend(['Team_A', 'Team_B'], loc = 4)

plt.show()
```

## BAR GRAPHS

```python
# team a
x1 = [2 ,5, 6, 7, 8]
y1 = [3 ,7, 4, 12 ,5]

# team b
x2 = [4 ,5, 8, 9, 8]
y2 = [3 ,5, 7, 12 ,6]
```

```python
plt.bar(x1, y1, color = 'g', linewidth = 4)
plt.bar(x2, y2, color = 'r', linewidth = 2)

plt.legend(['Team_A', 'Team_B'], loc = 2)
plt.title("BAR Plot")
plt.show()
```

```python
plt.barh(x1, y1, color = 'g', linewidth = 4)
plt.barh(x2, y2, color = 'r', linewidth = 2)

plt.legend(['Team_A', 'Team_B'], loc = 4)
plt.title("BAR Plot")
plt.show()
```

## Bubble Graph

```python
plt.figure(figsize = (15, 8))
x = [2, 5, 5, 6, 7, 8]
y = [3, 7, 7.2, 4, 12, 5]

plt.scatter(x, y, s = np.random.rand(6)*1000, c = np.random.rand(6))
plt.title('Bubble Graph')
plt.xlabel("x-axis values")
plt.ylabel("y-axis values")
plt.show()
```

## Pie chart

```python
plt.figure(figsize = (8, 8))
data = [3, 7, 4 ,12, 5]
label_name = ['3', '7', '4' ,'12', '5']
plt.pie(data, labels = label_name)
plt.show()
```
